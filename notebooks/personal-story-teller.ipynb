```python
def load_llm(model_name=model_name):
    
    
    # tokenizer = AutoTokenizer.from_pretrained(model_name,
    #                                           token = HF_TOKEN,
    #                                           )
    
    # model = AutoModelForCausalLM.from_pretrained(
    #     model_name,
    #     token = HF_TOKEN
    # )
    
    # pipe = pipeline(
    #     "text-generation",
    #     model=model,
    #     tokenizer=tokenizer,
    #     max_new_tokens=256,
    #     temperature=0.9,
    #     do_sample=True
    # )
        
    # llm = HuggingFacePipeline(pipeline=pipe)
    # llm_model = ChatHuggingFace(llm=llm)
    
    llm_model = OllamaLLM(
        model=model_name,
        temperature=0.7
    )
    
    return llm_model
```

```python
llm_model = load_llm()
```

```python
def generate_story(llm,topic):
    
    template = f"""Write an engaging and educational story about {topic} for beginners. 
            Use simple and clear language to explain basic concepts. 
            Include interesting facts and keep it friendly and encouraging. 
            The story should be around 200-300 words and end with a brief summary of what we learned. 
            Make it perfect for someone just starting to learn about this topic."""
                    
    prompt_template = PromptTemplate(
        template=template,
        input_variables=[]
        
    )
    
    chain_llm = prompt_template | llm | StrOutputParser()
    story = chain_llm.invoke({})
    
    return story
```

```python
topic = "the life cycle of butterflies"
story = generate_story(llm_model,topic)
print("Generated Story:\n", story)
```
